```python
import re
```

```python
# Define regex pattern to match age digits
age_pattern = r'\b(\d{1,2})\s*(?:years\*yo|s*old|yo|y.o|YO|year\s*old|)'

# Define function to extract age from text
def extract_age(patient_notes):
    age_match = re.search(age_pattern, patient_notes)
    if age_match:
        return int(age_match.group(1))
    else:
        return None

# Apply function to 'Text' column of dataframe
patient_notes['age'] = patient_notes['pn_history'].apply(extract_age)
```

```python
# define regex patterns for male and female
female_pattern = re.compile(r'\b(?:f|mrs|mrs.|ms|ms.|female|females|woman|womans|women|she|her)\b', re.IGNORECASE)
male_pattern = re.compile(r'\b(?:m|mr|mr.|male|males|man|he|his)\b', re.IGNORECASE)

# extract sex from text using regex
patient_notes['sex'] = patient_notes['pn_history'].apply(lambda x: 'male' if male_pattern.search(x) 
                                                         else 'female' if female_pattern.search(x)
                                                         else None)
```

```python
# define a regular expression pattern to match the sex
pattern = r'\b(m|mr|mr.|male|males|man|he|his|boy|fem[ia]le|f|mrs|mrs.|ms|ms.|female|females|woman|womans|women|she|her)\b'

# create a function to extract the sex using the regular expression pattern
def extract_sex(text):
    match = re.search(pattern, text, re.IGNORECASE)
    if match:
        return 'male' if match.group(1).lower() in ['male', 'males', 'mr', 'mr.', 'm', 'man', 'boy', 'he', 'his'] else 'female'
    else:
        return None
    
# apply the extract_sex function to the info column and create a new sex column in the dataframe
patient_notes['sex_1'] = patient_notes['pn_history'].apply(extract_sex)
```
