# Diabetes Health Indicators Analysis

Dataset source: https://www.kaggle.com/datasets/alexteboul/diabetes-health-indicators-dataset

The Behavioral Risk Factor Surveillance System (BRFSS) is a health-related telephone survey that is collected annually by the CDC. Each year, the survey collects responses from over 400,000 Americans on health-related risk behaviors, chronic health conditions, and the use of preventative services. In this dataset, a csv file available for the year 2015 was used.

### Diabetes
0: no diabetes or only during pregnancy, 1: prediabetes and 2: diabetes

### Features
The dataset consist of 21 features,

**HighBP or blood pressure (high)**
* Adults who have been told they have high blood pressure by a doctor, nurse, or other health professional
* 0: No, 1: Yes

**HighChol or cholesterol (high)**
* Have been told by a doctor, nurse or other health professional that your blood cholesterol is high
* 0: No, 1: Yes

**CholCheck**
* Cholesterol check within past five years
* 0: No, 1: Yes

**BMI (Body Mass Index)**

**Smoker**
* Have you smoked at least 100 cigarettes in your entire life? [Note: 5 packs = 100 cigarettes]
* 0: No, 1: Yes

**Stroke**
* Respondents that had a stroke
* 0: No, 1: Yes

**HeartDiseaseorAttack**
* Respondents that have ever reported having coronary heart disease (CHD) or myocardial infarction (MI)
* 0: No, 1: Yes

**PhysActivity**
* Adults who reported doing physical activity or exercise during the past 30 days other than their regular job
* 0: No, 1: Yes

**Fruit & Veggies**
* Consume Fruit 1 or more times per day
* Consume Vegetables 1 or more times per day
* 0: No, 1: Yes

**HvyAlcoholConsump**
* Heavy drinkers (adult men having more than 14 drinks per week and adult women having more than 7 drinks per week)
* 0: No, 1: Yes

**AnyHealthcare**
* Have any kind of health care coverage
* 0: No, 1: Yes

**NoDocbcCost**
* Need to see a doctor but could not because of cost
* 0: No, 1: Yes

**GenHlth**
* Respondents' health in scale of 1-5
* 1: Excellent
* 2: Very good
* 3: Good
* 4: Fair
* 5: Poor

**MentHlth**
* Days during the past 30 days was your mental health not good in 1-30 number of days

**PhysHlth**
* Days during the past 30 days was your physical health not good in number of 1-30 days

**DiffWalk**
* Have serious difficulty for walking or climbing stairs
* 0: No, 1: Yes

**Sex**
* 0: Female, 1: Male

**Age**
* Thirteen-level age category, 1-13
* 1: Age 18 to 24
* 2: Age 25 to 29
* 3: Age 30 to 34
* 4: Age 35 to 39
* 5: Age 40 to 44
* 6: Age 45 to 49
* 7: Age 50 to 54
* 8: Age 55 to 59
* 9: Age 60 to 64
* 10: Age 65 to 69
* 11: Age 70 to 74
* 12: Age 75 to 79

**Education**
* Highest grade or year of school completed
* 1: Never attended school or only kindergarten
* 2: Grades 1 through 8 (Elementary)
* 3: Grades 9 through 11 (Some high school)
* 4: Grade 12 or GED (High school graduate)
* 5: College 1 year to 3 years (Some college or technical school)
* 6: College 4 years or more (College graduate)

**Income**
* Annual household income (in USD) from all sources in scale of 1-8
* 1: Less than 10,000
* 2: Less than 15,000
* 3: Less than 20,000

## What risk factors are most predictive of diabetes risk?

From heat map above, it can be conluded that the risk factors that best predict the risk of diabetes are general health, blood pressure level, BMI, difficulty to walk and cholesterol level.

### BMI

```python
plt.figure(figsize=(15,5))

sns.histplot(x=df['BMI'], color='b') # BMI of all respondents
sns.histplot(x=df.BMI[df.Diabetes_012 == 2], color='r') # BMI of respondents with type 2 diabetes

plt.show()
```

#### BMI Levels

* **Underweight range**: BMI less than 18.5
* **Healthy weight range**: BMI of 18.5 to <25
* **Overweight range**: BMI of 25.0 to <30
* **Obesity range**: BMI of 30.0 or higher

#### Obesity is frequently subdivided into categories:

* Class 1: BMI of 30 to < 35
* Class 2: BMI of 35 to < 40
* Class 3: BMI of 40 or higher. Class 3 obesity is sometimes categorized as “severe” obesity.

Source: https://www.cdc.gov/obesity/basics/adult-defining.html#:~:text=If%20your%20BMI%20is%20less,falls%20within%20the%20obesity%20range.

```python
def data_label(BMI):
    if BMI['BMI'] < 18.5:
        return 'underweight'
    elif BMI['BMI'] >= 18.5 and BMI['BMI'] <25.0:
        return 'healthy'
    elif BMI['BMI'] >= 25.0 and BMI['BMI'] < 30.0:
        return 'overweight'
    elif BMI['BMI'] >= 30.0 and BMI['BMI'] < 35.0:
        return 'obesity class 1'
    elif BMI['BMI'] >= 35.0 and BMI['BMI'] < 40.0:
        return 'obesity class 2'
    else:
        return 'obesity class 3'
```

```python
df['BMILevel'] = df.apply(lambda BMI: data_label(BMI), axis=1)
```

```python
df.head()
```

```python
df['BMILevel'].value_counts()
```

```python
plt.figure(figsize=(13, 7))

ax = sns.countplot(x='BMILevel', hue='Diabetes_012', data=df)
for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

**-** Respondents with underweight and healthy BMI levels had the low number of type 2 diabetes.

**-** Meanwhile, respondents with a BMI level above healthy level, have a higher number of type 2 diabetes.

### HighBP and HighChol

```python
def new_feature(HighBPChol):
    if HighBPChol['HighBP'] == 1.0 and HighBPChol['HighChol'] == 1.0:
        return 'HighBPChol'
    if HighBPChol['HighBP'] == 1.0 and HighBPChol['HighChol'] == 0.0:
        return 'HighBP'
    if HighBPChol['HighBP'] == 0.0 and HighBPChol['HighChol'] == 1.0:
        return 'HighChol'
    else:
        return 'LowBPChol'
```

```python
df['HighBPChol'] = df.apply(lambda HighBPChol: new_feature(HighBPChol), axis=1)
```

```python
plt.figure(figsize=(10,5))

ax = sns.countplot(x='HighBPChol', hue='Diabetes_012', data=df)
for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

**-** From the **35,097** respondents with type 2 diabetes, **21%** of them had high blood pressure and the other **12%** had high cholesterol. While **54%** of respondents have both high blood pressure and cholesterol.

**-** More than **87%** of respondents who suffer from type 2 diabetes have one or both of the above conditions.

**-** Respondents with high blood pressure and cholesterol were **more likely** to have type 2 diabetes than respondents with high blood pressure alone (almost 2 times) or high cholesterol alone (almost 3 times).
